def initialize_inventory(products):

````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
````

````python
def get_customer_orders(inventory):
    customer_orders = []
    
    while True:
        product = input("Enter product name (or type 'done' to finish): ")

        if product.lower() == "done":
            break

        if product not in inventory:
            print("Product does not exist. Try again.")
        elif inventory[product] == 0:
            print("Product is out of stock.")
        else:
            customer_orders.append(product)
            inventory[product] -= 1
            print(product + " added to the order.")

    return customer_orders
````

````python
def calculate_total_price(customer_orders):
    total_price = 0

    for product in customer_orders:
        valid_price = False

        while not valid_price:
            try:
                price = int(input(f"Enter price for {product}: "))

                if price < 0:
                    print("Price cannot be negative.")
                else:
                    total_price += price
                    valid_price = True

            except ValueError:
                print("Please enter a valid number.")

    return total_price
````

````python
products = ["apple", "banana", "orange", "grapes"]

# Initialize inventory
inventory = initialize_inventory(products)

# Get orders
customer_orders = get_customer_orders(inventory)

# Calculate total price
total = calculate_total_price(customer_orders)

# Print results
print("Customer ordered:", customer_orders)
print("Total price:", total)
print("Updated inventory:", inventory)
````
